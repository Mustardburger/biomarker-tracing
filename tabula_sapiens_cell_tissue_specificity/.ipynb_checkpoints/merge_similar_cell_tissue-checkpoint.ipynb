# Feature selection

## From highly variable genes, irrespective of whether these include Olink proteins

```python
def find_connected_components(graph):
    """
    """
    visited = set()
    components = []
    
    def dfs(node, component):
        """Depth-first search to explore a connected component."""
        visited.add(node)
        component.append(node)
        
        # Explore all neighbors
        for neighbor in graph.get(node, []):
            if neighbor not in visited:
                dfs(neighbor, component)
    
    # Find all connected components
    for node in graph:
        if node not in visited:
            component = []
            dfs(node, component)
            components.append(component)
    
    return components
```

```python
from scipy.cluster.hierarchy import linkage, fcluster

# atlas_smal = tot_data.loc[:, atlas.columns]

# Calculate the correlation matrix
corr_thres = 0.8
corr_mat = atlas_smal.corr()
corr_mat_arr = corr_mat.to_numpy()
true_ind = list(np.column_stack(np.where(corr_mat_arr > corr_thres)))
visited = set()

# Very manual: loop through the columns and find any combination with correlation above threshold
print("Creating adjacency list of correlated elements")
d = {i: [] for i in range(corr_mat.shape[0])}
for ind in true_ind:
    i1, i2 = ind
    if i1 == i2 or f"{i1}-{i2}" in visited: continue
    else:
        visited.add(f"{i1}-{i2}")
        visited.add(f"{i2}-{i1}")
        d[i1].append(i2)
        d[i2].append(i1)

# Find connected elements
print("Finding connected components in d")
groups = find_connected_components(d)

# Aggregating similar groups together using some func
print("Creating new dataframe")
atlas_smal_arr = atlas_smal.to_numpy()
new_arr = np.zeros((atlas_smal.shape[0], len(groups)))
col_names = [""]*len(groups)
func = np.mean                   
for i, g in enumerate(groups):
    new_col = func(atlas_smal_arr[:, list(g)], axis=1)
    new_arr[:, i] = new_col
    if len(g) > 1: col_names[i] = f"merged_ind_{i}"
    else: 
        col_name = list(g)[0]
        col_names[i] = corr_mat.columns[col_name]

atlas_smal_merged = pd.DataFrame(new_arr, columns=col_names, index=atlas_smal.index)
print(f"New dataframe shape = {atlas_smal_merged.shape}")
```

```python
### Post-merge quality control
# Plotting the correlation matrix
data = atlas_smal_merged
corr = data.corr()
tmp_zzz = atlas_smal.copy()
plt.figure()
sns.clustermap(corr >= corr_thres, xticklabels=False, yticklabels=False, figsize=(8, 8), vmin=0.0, vmax=1.0)
plt.show()
```

```python
# Check names if merged groups whether they represent any biologically similar cell-tissues
group_ind = 63
group = groups[group_ind]
#tmp = atlas_smal.copy()

group_ele_names = tmp_zzz.columns[list(group)].tolist()
cells = [i.split("_")[1] for i in group_ele_names]
tissues = [i.split("_")[0] for i in group_ele_names]
tmp_df = pd.DataFrame({"group_cell": cells, "group_tissue": tissues})
tmp_df
```

```python
# These merged features are independent of Olink proteins

# Now, using these merged features, generate Olink-specific gene expression matrix
olink_atlas = atlas.loc[olink_prot["gene"].tolist()]
atlas_smal_arr = olink_atlas.to_numpy()
new_arr = np.zeros((olink_atlas.shape[0], len(groups)))
col_names = [""]*len(groups)
func = np.mean                   
for i, g in enumerate(groups):
    new_col = func(atlas_smal_arr[:, list(g)], axis=1)
    new_arr[:, i] = new_col
    if len(g) > 1: col_names[i] = f"merged_ind_{i}"
    else: 
        col_name = list(g)[0]
        col_names[i] = corr_mat.columns[col_name]
olink_atlas_merged = pd.DataFrame(new_arr, columns=col_names, index=olink_atlas.index)

# Plotting the correlation matrix
data = olink_atlas_merged
corr = data.corr()
plt.figure()
sns.clustermap(corr >= 0.85, xticklabels=False, yticklabels=False, figsize=(8, 8), vmin=0.0, vmax=1.0)
plt.show()
```

## Single Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

# Prep the labels (whether a protein in Olink or not)
all_genes = X_df.index.tolist()
gene_labels = [i in olink_prot["gene"].tolist() for i in all_genes]
gene_label_df = pd.DataFrame({"gene": all_genes, "label": gene_labels}).set_index("gene")
gene_labels_ordered = X_df.merge(gene_label_df, right_index=True, left_index=True)["label"].tolist()

features = X_df.columns.tolist()
model_coefs = []
for feature in tqdm.tqdm(features):
    
    model = LogisticRegression()
    model.fit(X_df[feature].to_frame(), gene_labels_ordered)
    model_coefs.append(model.coef_.item())

coef_df_sin = pd.DataFrame({"coeff_sin": model_coefs, "cell_tissue": features})
coef_df_sin.sort_values(by="coeff_sin", ascending=False).head(10)
```

```python
coef_df = coef_df_mul.merge(coef_df_sin, on="cell_tissue", how="left")
coef_df = pd.concat([
    coef_df,
    coef_df[["coeff_mul", "coeff_sin"]].rank(ascending=False).astype(int).rename(columns={"coeff_mul": "coeff_mul_rank", "coeff_sin": "coeff_sin_rank"})
], axis=1)
coef_df["rank_diff"] = np.abs(coef_df["coeff_mul_rank"] - coef_df["coeff_sin_rank"])
plt.figure()
sns.jointplot(coef_df, x="coeff_sin", y="coeff_mul")
plt.show()
```

## Recursive feature elimination

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.feature_selection import RFE
from sklearn.svm import SVC, LinearSVC

# Prep the labels (whether a protein in Olink or not)
all_genes = X_df.index.tolist()
gene_labels = [i in olink_prot["gene"].tolist() for i in all_genes]
gene_label_df = pd.DataFrame({"gene": all_genes, "label": gene_labels}).set_index("gene")
gene_labels_ordered = X_df.merge(gene_label_df, right_index=True, left_index=True)["label"].tolist()

types = ["linear_svc", "logreg"]
models = [
    LinearSVC(max_iter=5000, penalty="l1", dual=False), LogisticRegression()
]
rfe_df = [X_df.columns]
for typ, model in zip(types, models):
    print(f"Working on {typ}")
    rfe = RFE(model, n_features_to_select=15)
    rfe.fit(X_df, gene_labels_ordered)
    supports = rfe.support_
    rankings = rfe.ranking_
    rfe_df.append(supports)
    rfe_df.append(rankings)

cols = ["cell_tissue"] + [f"{a}_{i}" for i in types for a in ["rfe_chosen", "ranking"]]
rfe_df = pd.DataFrame(rfe_df).T
rfe_df.columns = cols
coef_rfe_df = coef_df.merge(rfe_df, on="cell_tissue", how="left")
```

```python
coef_rfe_df["all"] = coef_rfe_df["rfe_chosen_linear_svc"] & coef_rfe_df["rfe_chosen_logreg"]
coef_rfe_df.to_csv("/sc/arion/projects/DiseaseGeneCell/Huang_lab_project/BioResNetwork/Phuc/projects/Alzheimer/human_atlas/sub_projects/plasma_proteome/results/classify_cel-tis_olink_non-olink_genes.tsv", sep="\t", index=False)
```

```python
plt.figure()
sns.boxplot(data=coef_rfe_df, x="all", y="coeff")
sns.stripplot(data=coef_rfe_df, x="all", y="coeff")
plt.show()
```

```python
atlas_used = pd.read_csv("/sc/arion/projects/DiseaseGeneCell/Huang_lab_project/BioResNetwork/Phuc/projects/Alzheimer/human_atlas/sub_projects/plasma_proteome/results/atlas_smal_merged_graph-merged.tsv", sep="\t").set_index("gene")

from sklearn.preprocessing import StandardScaler, PowerTransformer, QuantileTransformer, RobustScaler

norm_method = "zscore"
if norm_method == "yeo-johnson":
    obj = PowerTransformer(method="yeo-johnson")
elif norm_method == "zscore":
    obj = StandardScaler()
elif norm_method == "quantile":
    obj = QuantileTransformer(output_distribution='normal')
elif norm_method == "robust":
    obj = RobustScaler()

X = obj.fit_transform(atlas_used.to_numpy())
X_df = pd.DataFrame(X, columns=atlas_used.columns, index=atlas_used.index)
```

```python
perc = 0.1

num_zero_before_trans = np.sum(atlas_used == 0, axis=0).to_frame().rename(columns={0: "num_zeros"}).reset_index(names="cell_tissue")
perc_zero_before_trans = np.mean(atlas_used == 0, axis=0).to_frame().rename(columns={0: "perc_zeros"}).reset_index(names="cell_tissue")
quantile_val_after_trans = X_df.quantile(perc).to_frame().rename(columns={perc: f"val_perc_{perc}"}).reset_index(names="cell_tissue")
final_df = (
    rfe_df
    .merge(num_zero_before_trans, on="cell_tissue", how="left")
    .merge(perc_zero_before_trans, on="cell_tissue", how="left")
    .merge(quantile_val_after_trans, on="cell_tissue", how="left")
)

plt.figure()
sns.boxplot(data=final_df, x="rfe_chosen", y="perc_zeros")
sns.stripplot(data=final_df, x="rfe_chosen", y="perc_zeros")
plt.show()

plt.figure()
sns.boxplot(data=final_df, x="rfe_chosen", y=f"val_perc_{perc}")
sns.stripplot(data=final_df, x="rfe_chosen", y=f"val_perc_{perc}")
plt.show()
```
