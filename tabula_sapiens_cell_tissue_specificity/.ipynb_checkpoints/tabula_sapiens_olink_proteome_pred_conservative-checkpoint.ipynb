## Graph-based correlation

```python
def find_connected_components(graph):
    """
    """
    visited = set()
    components = []
    
    def dfs(node, component):
        """Depth-first search to explore a connected component."""
        visited.add(node)
        component.append(node)
        
        # Explore all neighbors
        for neighbor in graph.get(node, []):
            if neighbor not in visited:
                dfs(neighbor, component)
    
    # Find all connected components
    for node in graph:
        if node not in visited:
            component = []
            dfs(node, component)
            components.append(component)
    
    return components
```

```python
from scipy.cluster.hierarchy import linkage, fcluster

# atlas_smal = tot_data.loc[:, atlas.columns]

# Calculate the correlation matrix
corr_thres = 0.7
corr_mat = atlas_smal.corr()
corr_mat_arr = corr_mat.to_numpy()
true_ind = list(np.column_stack(np.where(corr_mat_arr > corr_thres)))
visited = set()

# Very manual: loop through the columns and find any combination with correlation above threshold
print("Creating adjacency list of correlated elements")
d = {i: [] for i in range(corr_mat.shape[0])}
for ind in true_ind:
    i1, i2 = ind
    if i1 == i2 or f"{i1}-{i2}" in visited: continue
    else:
        visited.add(f"{i1}-{i2}")
        visited.add(f"{i2}-{i1}")
        d[i1].append(i2)
        d[i2].append(i1)

# Find connected elements
print("Finding connected components in d")
groups = find_connected_components(d)

# Aggregating similar groups together using some func
print("Creating new dataframe")
atlas_smal_arr = atlas_smal.to_numpy()
new_arr = np.zeros((atlas_smal.shape[0], len(groups)))
col_names = [""]*len(groups)
func = np.mean                   
for i, g in enumerate(groups):
    new_col = func(atlas_smal_arr[:, list(g)], axis=1)
    new_arr[:, i] = new_col
    if len(g) > 1: col_names[i] = f"merged_ind_{i}"
    else: 
        col_name = list(g)[0]
        col_names[i] = corr_mat.columns[col_name]

atlas_smal_merged = pd.DataFrame(new_arr, columns=col_names, index=atlas_smal.index)
print(f"New dataframe shape = {atlas_smal_merged.shape}")
```

## Hierarchical clustering

```python
from scipy.cluster import hierarchy
from scipy.spatial.distance import squareform
from scipy.stats import spearmanr

# Code snippet from https://scikit-learn.org/stable/auto_examples/inspection/plot_permutation_importance_multicollinear.html

X = atlas_smal
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 8))
corr = spearmanr(X).correlation

# Ensure the correlation matrix is symmetric
corr = (corr + corr.T) / 2
np.fill_diagonal(corr, 1)

# We convert the correlation matrix to a distance matrix before performing
# hierarchical clustering using Ward's linkage.
distance_matrix = 1 - np.abs(corr)
dist_linkage = hierarchy.ward(squareform(distance_matrix))
dendro = hierarchy.dendrogram(
    dist_linkage, ax=ax1, leaf_rotation=90
)
dendro_idx = np.arange(0, len(dendro["ivl"]))

ax2.imshow(corr[dendro["leaves"], :][:, dendro["leaves"]])
#ax2.set_xticks(dendro_idx)
#ax2.set_yticks(dendro_idx)
#ax2.set_xticklabels(dendro["ivl"], rotation="vertical")
#ax2.set_yticklabels(dendro["ivl"])
_ = fig.tight_layout()
```

```python
from collections import defaultdict

score = 0.5
cluster_ids = hierarchy.fcluster(dist_linkage, score, criterion="distance")
cluster_id_to_feature_ids = defaultdict(list)
for idx, cluster_id in enumerate(cluster_ids):
    cluster_id_to_feature_ids[cluster_id].append(idx)
selected_features = [v[0] for v in cluster_id_to_feature_ids.values()]
selected_features_names = X.columns[selected_features]
print(len(selected_features_names))
```

# Data norm

```python
### Z-score: this thread shows the necessity for z-score when doing LASSO: 
# https://stats.stackexchange.com/questions/86434/is-standardisation-before-lasso-really-necessary
from sklearn.preprocessing import StandardScaler, PowerTransformer, QuantileTransformer, RobustScaler, FunctionTransformer

# Create an atlas of both highly variable genes and Olink proteins
atlas_smal = atlas.loc[var_features.head(N)["var.features"].tolist()]
all_features = list(set(atlas_smal.index.tolist()).union(set(olink_prot["gene"].tolist())))
print(len(all_features))

# Now, using these features, generate gene expression matrix
olink_hvg_atlas = atlas.loc[all_features]
atlas_smal_arr = olink_hvg_atlas.to_numpy()
new_arr = np.zeros((olink_hvg_atlas.shape[0], len(groups)))
col_names = [""]*len(groups)
func = np.mean                   
for i, g in enumerate(groups):
    new_col = func(atlas_smal_arr[:, list(g)], axis=1)
    new_arr[:, i] = new_col
    if len(g) > 1: col_names[i] = f"merged_ind_{i}"
    else: 
        col_name = list(g)[0]
        col_names[i] = corr_mat.columns[col_name]
atlas_smal = pd.DataFrame(new_arr, columns=col_names, index=olink_hvg_atlas.index)

norm_method = "zscore"
if norm_method == "yeo-johnson":
    obj = PowerTransformer(method="yeo-johnson")
elif norm_method == "zscore":
    obj = StandardScaler()
elif norm_method == "quantile":
    obj = QuantileTransformer(output_distribution='normal')
elif norm_method == "robust":
    obj = RobustScaler()
elif norm_method == "identity":
    obj = FunctionTransformer()

X = obj.fit_transform(atlas_smal.to_numpy())
X_df = pd.DataFrame(X, columns=atlas_smal.columns, index=atlas_smal.index)

# Check whether merged columns have smaller variance than individual cell-tissue columns
var_col = X_df.max(axis=0).to_frame().reset_index().rename(columns={"index": "cell-tissue", 0: "variance"})
var_col["col_type"] = var_col["cell-tissue"].apply(lambda x: "merged" if "merged" in x else "indivi")
plt.figure()
sns.boxplot(data=var_col, x="col_type", y="variance")
sns.stripplot(data=var_col, x="col_type", y="variance")
plt.show()
```

## Single Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

# Prep the labels (whether a protein in Olink or not)
all_genes = X_df.index.tolist()
gene_labels = [i in olink_prot["gene"].tolist() for i in all_genes]
gene_label_df = pd.DataFrame({"gene": all_genes, "label": gene_labels}).set_index("gene")
gene_labels_ordered = X_df.merge(gene_label_df, right_index=True, left_index=True)["label"].tolist()

features = X_df.columns.tolist()
model_coefs = []
for feature in tqdm.tqdm(features):
    
    model = LogisticRegression()
    model.fit(X_df[feature].to_frame(), gene_labels_ordered)
    model_coefs.append(model.coef_.item())

coef_df_sin = pd.DataFrame({"coeff_sin": model_coefs, "cell_tissue": features})
coef_df_sin.sort_values(by="coeff_sin", ascending=False).head(10)
```

```python
coef_df = coef_df_mul.merge(coef_df_sin, on="cell_tissue", how="left")
coef_df = pd.concat([
    coef_df,
    coef_df[["coeff_mul", "coeff_sin"]].rank(ascending=False).astype(int).rename(columns={"coeff_mul": "coeff_mul_rank", "coeff_sin": "coeff_sin_rank"})
], axis=1)
coef_df["rank_diff"] = np.abs(coef_df["coeff_mul_rank"] - coef_df["coeff_sin_rank"])
plt.figure()
sns.jointplot(coef_df, x="coeff_sin", y="coeff_mul")
plt.show()
```

## Recursive feature elimination

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.feature_selection import RFE
from sklearn.svm import SVC, LinearSVC

# Prep the labels (whether a protein in Olink or not)
all_genes = X_df.index.tolist()
gene_labels = [i in olink_prot["gene"].tolist() for i in all_genes]
gene_label_df = pd.DataFrame({"gene": all_genes, "label": gene_labels}).set_index("gene")
gene_labels_ordered = X_df.merge(gene_label_df, right_index=True, left_index=True)["label"].tolist()

types = ["linear_svc", "logreg"]
models = [
    LinearSVC(max_iter=5000, penalty="l1", dual=False), LogisticRegression()
]
rfe_df = [X_df.columns]
for typ, model in zip(types, models):
    print(f"Working on {typ}")
    rfe = RFE(model, n_features_to_select=15)
    rfe.fit(X_df, gene_labels_ordered)
    supports = rfe.support_
    rankings = rfe.ranking_
    rfe_df.append(supports)
    rfe_df.append(rankings)

cols = ["cell_tissue"] + [f"{a}_{i}" for i in types for a in ["rfe_chosen", "ranking"]]
rfe_df = pd.DataFrame(rfe_df).T
rfe_df.columns = cols
#coef_rfe_df = coef_df.merge(rfe_df, on="cell_tissue", how="left")
```
